```python
def trap_mf(x,a,b,c,d):
    if x<=a or x>d:
        return 0.0
    elif a<x<=b:
        return (x-a)/(b-a)
    elif b<x<=c:
        return 1.0
    elif c<x<=d:
        return (d-x)/(d-c)
```

```python
def bp_low(x):
    if x<80:
        return 1.0
    else:
        return trap_mf(x,80,80,80,100)
```

```python
def bp_normal(x):
    return trap_mf(x,80,100,120,140)
```

```python
def bp_high(x):
    if x>140:
        return 1.0
    else:
        return trap_mf(x,120,120,120,140)
```

```python
def temp_normal(x):
    return trap_mf(x,34,36,38,40)

def temp_high(x):
    if x>40:
        return 1
    else:
        return trap_mf(x,38,38,38,40)
```

```python
t=float(input("Enter the temperature:"))
b=float(input("Enter the blood pressure:"))

bp_high_m=bp_high(b)
bp_normal_m=bp_normal(b)
bp_low_m=bp_low(b)

temp_high_m=temp_high(t)
temp_normal_m=temp_normal(t)
```
