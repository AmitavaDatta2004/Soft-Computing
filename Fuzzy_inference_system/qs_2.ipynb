```python
# QS 2
import numpy as np

def bell(x,a,b,c):
    return 1/(1+np.abs(((x-c)/a)**(2*b)))
```

```python
x=float(input("Enter the a age of the person:"))
```

```python
x
```

```python
y=bell(x,20,2,0)
o=bell(x,30,3,100)
```

```python
more_or_less_young=y**0.5
print(more_or_less_young)
```

```python
not_young_and_not_old=min((1-y),(1-o))
print(not_young_and_not_old)
```

```python
young_but_not_too_young=min(y,(1-y**2))
print(young_but_not_too_young)
```
