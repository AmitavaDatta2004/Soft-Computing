```python
T={ 16:0.4,
    18:0.8,
    20:1.0,
    22:1.0,
    24:0.8,
    26:0.5
    }

H={ 0:0.2,
    20:0.8,
    40:1.0,
    60:0.6,
    80:0.2
    }
```

```python
th_t=float(input("Enter the threshhold of temperature (0-1) :"))
th_h=float(input("Enter the threshold of humidity (0-1 :)"))
```

```python
th_t
```

```python
th_h
```

```python
for i in T:
    for j in H:
        if T[i] >= th_t and H[j] >= th_h:
            print(f"{i},{j}={max(T[i],H[j])}")
```
