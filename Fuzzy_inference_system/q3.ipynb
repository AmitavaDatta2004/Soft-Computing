```python
# defuzzy ficatication methods

elements=[]
membership=[]

n=int(input("Enter the no of elements"))

for i in range(n):
    elements.append(float(input(f"Enter the {i+1} element: ")))
    membership.append(float(input(f"Enter the membership value of {i+1} element: ")))
```

```python
# centroid method (weighted sum)

ws=0
s=0
for i in range(n):
    ws+=(elements[i]*membership[i])
    s+=(membership[i])
```

```python
# bisector method

total=s
cumulative=0

for i in range(n):
    cumulative+=membership[i]
    if cumulative>=total/2:
        boa=elements[i]
        break

```

```python
# mean of maximum
max_m=max(membership)
print(max_m)
```

```python
max_elements=[]
for i in range(n):
    if membership[i]==max_m:
        max_elements.append(elements[i])

max_elements_sum=sum(max_elements)
mom=max_elements_sum/len(max_elements)
```

```python
# smallest of maximum

som = min(max_elements)
```
