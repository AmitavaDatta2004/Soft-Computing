```python
def tri_mf(x,a,b,c):
    if x<=a or x>=c:
        return 0
    if a<x<=b:
        return (x-a)/(b-a)
    if b<x<c:
        return (c-x)/(c-b)
```

```python
import numpy as np
x=np.linspace(0,10,1000)
result_tri=[]
for i in x:
    result_tri.append(tri_mf(i,2,5,8))
```

```python
import matplotlib.pyplot as plt
plt.plot(x,result_tri)
```

```python
def trap_mf(x,a,b,c,d):
    if x<=a or x>=d:
        return 0
    if a<x<=b:
        return (x-a)/(b-a)
    if b<x<=c:
        return 1
    if c<x<d:
        return (d-x)/(d-c)
```

```python
result_trap=[]
for i in x:
    result_trap.append(trap_mf(i,2,4,6,8))
```

```python
plt.plot(x,result_trap)
```

```python
import math
def gauss_mf(x,c,sigma):
    return math.exp(-((x-c)**2)/(sigma**2))
```

```python
result_gauss=[]
for i in x:
    result_gauss.append(gauss_mf(i,5,1))
```

```python
plt.plot(x,result_gauss)
```

```python
def bell_mf(x,a,b,c):
    return 1/(1+(np.abs((x-c)/a))**(2*b))
```

```python
result_bell=[]
for i in x:
    result_bell.append(bell_mf(i,2,3,5))
```

```python
plt.plot(x,result_bell)
```

```python
def sigmoid_mf(x,a,c):
    return 1/(1+math.exp(-(a*(x-c))))
```

```python
result_sigmoid=[]
for i in x:
    result_sigmoid.append(sigmoid_mf(i,2,5))
```

```python
plt.plot(x,result_sigmoid)
```

```python
M1=[[0.5,0.7,0.2],[0.8,0.6,0.9]]
M2=[[0.6,0.3],[0.9,0.4],[0.5,0.8]]

def max_min(R,S):
    final=[]
    for i in range(len(R)):
        row=[]
        for j in range(len(S[0])):
            element=[]
            for k in range(len(S)):
                element.append(min(R[i][k],S[k][j]))
            row.append(max(element))
        final.append(row)
    return final
```

```python
max_min(M1,M2)
```

```python
def max_product(R,S):
    final=[]
    for i in range(len(R)):
        row=[]
        for j in range(len(S[0])):
            product=[]
            for k in range(len(S)):
                product.append(R[i][k]*S[k][j])
            row.append(max(product))
        final.append(row)
    return final
```

```python
max_product(M1,M2)
```
