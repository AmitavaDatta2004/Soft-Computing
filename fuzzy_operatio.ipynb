```python
def tri(x,a,b,c):
    if x<=a or x>=c :
        return 0
    if a<x<=b:
        return (x-a)/(b-a)
    if b<x<c:
        return (c-x)/(c-b)
```

```python
import numpy as np
x=np.linspace(0,10,1000)
```

```python
results1=[]
results2=[]

for i in x:
    results1.append(tri(i,2,4,6))
    results2.append(tri(i,4,6,8))
```

```python
union=[]
intersection=[]
complement_1=[]
complement_2=[]

for i in range(len(x)):
    union.append(max(results1[i],results2[i]))
    intersection.append(min(results1[i],results2[i]))
    complement_1.append(1-results1[i])
    complement_2.append(1-results2[i])
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(x,results1)
plt.plot(x,results2)
```

```python
plt.plot(x,union)
plt.plot(x,intersection)
```

```python
plt.plot(x,results1)
plt.plot(x,complement_1)
```

```python
plt.plot(x,results2)
plt.plot(x,complement_2)
```

```python
sum=[]
difference=[]
for i in range(len(x)):
    sum.append(results1[i]+results2[i]-(results1[i]*results2[i]))
    difference.append(results1[i]*(1-results2[i]))
```

```python
plt.plot(x,sum)
```

```python
plt.plot(x,difference)
```
