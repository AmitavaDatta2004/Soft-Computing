```python
def tri(x,a,b,c):
    if x<=a or x>=c :
        return 0
    if a<x<=b:
        return (x-a)/(b-a)
    if b<x<c:
        return (c-x)/(c-b)
```

```python
import numpy as np
x=np.linspace(0,10,1000)
```

```python
results1=[]
results2=[]

for i in x:
    results1.append(tri(i,2,4,6))
    results2.append(tri(i,4,6,8))
```

```python
union=[]
intersection=[]
complement_1=[]
complement_2=[]

for i in range(len(x)):
    union.append(max(results1[i],results2[i]))
    intersection.append(min(results1[i],results2[i]))
    complement_1.append(1-results1[i])
    complement_2.append(1-results2[i])
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10, 6))
plt.plot(x, results1, label='Tri(x, 2, 4, 6)')
plt.plot(x, results2, label='Tri(x, 4, 6, 8)')
plt.title('Triangular Membership Functions')
plt.xlabel('x')
plt.ylabel('Membership Value')
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(x, union, label='Union')
plt.title('Fuzzy Set Union')
plt.xlabel('x')
plt.ylabel('Membership Value')
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(x, intersection, label='Intersection')
plt.title('Fuzzy Set Intersection')
plt.xlabel('x')
plt.ylabel('Membership Value')
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(x, results1, label='Original Set 1')
plt.plot(x, complement_1, label='Complement of Set 1')
plt.title('Fuzzy Set and Its Complement (Set 1)')
plt.xlabel('x')
plt.ylabel('Membership Value')
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(x, results2, label='Original Set 2')
plt.plot(x, complement_2, label='Complement of Set 2')
plt.title('Fuzzy Set and Its Complement (Set 2)')
plt.xlabel('x')
plt.ylabel('Membership Value')
plt.legend()
plt.grid(True)
plt.show()
```

```python
sum=[]
difference=[]
for i in range(len(x)):
    sum.append(results1[i]+results2[i]-(results1[i]*results2[i]))
    difference.append(results1[i]*(1-results2[i]))
```

```python
plt.figure(figsize=(10, 6))
plt.plot(x, sum)
plt.title('Fuzzy Algebraic Sum')
plt.xlabel('x')
plt.ylabel('Membership Value')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(x, difference)
plt.title('Fuzzy Difference (Set 1 AND NOT Set 2)')
plt.xlabel('x')
plt.ylabel('Membership Value')
plt.grid(True)
plt.show()
```
