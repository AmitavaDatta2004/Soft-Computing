```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
x=np.linspace(0,10,1000)
```

```python
print(x)      
```

```python
def tri(x,a,b,c):
    if x<=a or x>=c:
        return 0
    if a<x<=b:
        return (x-a)/(b-a)
    if b<x<c:
        return (c-x)/(c-b)
```

```python
result=[]
for i in x:
    result.append(tri(i,2,4,6))
```

```python
print(tri(4.5,4,5,6))
```

```python
result
```

```python
plt.plot(x,result)
```
