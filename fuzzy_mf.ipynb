```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
x=np.linspace(0,10,1000)
```

```python
print(x)      
```

```python
def tri(x,a,b,c):
    if x<=a or x>=c:
        return 0
    if a<x<=b:
        return (x-a)/(b-a)
    if b<x<c:
        return (c-x)/(c-b)
```

```python
result_tri=[]
for i in x:
    result_tri.append(tri(i,2,4,6))
```

```python
print(tri(4.5,4,5,6))
```

```python
result_tri
```

```python
plt.plot(x,result_tri)
```

```python
def trap(x,a,b,c,d):
    if x<=a or x >=d:
        return 0
    if a<x<=b:
        return (x-a)/(b-a)
    if b<x<=c:
        return 1
    if c<x<d:
        return (d-x)/(d-c)
```

```python
result_trap=[]
for i in x:
    result_trap.append(trap(i,2,4,6,8))
```

```python
result_trap
```

```python
plt.plot(x,result_trap)
```

```python
def gauss(x,c,sigma):
    return math.exp(-((x-c)**2)/(2*(sigma**2)))
```

```python
result_sigma=[]
for i in x:
    result_sigma.append(gauss(i,5,1))
```

```python
plt.plot(x,result_sigma)
```

```python
def bell(x,a,b,c):
    return 1/(1+((x-c)/a)**(2*b))
```

```python
result_bell=[]
for i in x:
    result_bell.append(bell(i,2,3,5))
```

```python
plt.plot(x,result_bell)
```
