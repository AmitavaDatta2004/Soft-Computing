```python
import math
def sigmoid(x):
    return 1/(1+math.exp(-x))
```

```python
def train(inputs,target):
    w1,w2,b=0.5,0.5,1
    lr=0.05
    epochs=10000
    for epoch in range(epochs):
        for i in range(len(inputs)):
            x1,x2=inputs[i]
            output=sigmoid(w1*x1+w2*x2+b)
            error=target[i]-output

            w1=w1+lr*error*x1
            w2=w2+lr*error*x2
            b=b+lr*error

    return w1,w2,b
```

```python
inputs=[[0,0],[0,1],[1,0],[1,1]]
or_out=[0,1,1,1]
and_out=[0,0,0,1]
xor_out=[0,1,0,1]
```

```python
w1,w2,b=train(inputs,or_out)
```
