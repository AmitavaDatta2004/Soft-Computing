```python
# backpropagation xor
inputs=[[0,0],[0,1],[1,0],[1,1]]
xor_out=[0,1,1,0]

import math
def sigmoid(x):
    return 1/(1+math.exp(-x))

```

```python
epochs=100000

w1,w2,w3,w4 = 0.5,0.5,0.5,0.5
b1 , b2 = 1,1

w5,w6=0.5,0.5
b3=1

lr=0.05

for epoch in range(epochs):
    for i in range(len(inputs)):
        x1,x2=inputs[i]

        h1=sigmoid(w1*x1+w2*x2+b1)
        h2=sigmoid(w3*x1+w4*x2+b2)

        o=sigmoid(h1*w5+h2*w6+b3)

        error=xor_out[i]-o

        delta_o=error*o*(1-o)

        delta_h1=delta_o*w5*h1*(1-h1)
        delta_h2=delta_o*w6*h2*(1-h2)

        w1=w1+delta_h1*lr*x1
        w2=w2+delta_h1*lr*x1
        b1=b1+delta_h1*lr

        w3=w3+delta_h2*lr*x2
        w4=w4+delta_h2*lr*x2
        b2=b2+delta_h2*lr

        w5=w5+delta_o*lr*h1
        w6=w6+delta_o*lr*h2
        b3=b3+delta_o*lr


for i in range(len(inputs)):
    x1,x2=inputs[i]
    h1 = sigmoid(w1*x1 + w2*x2 + b1)
    h2 = sigmoid(w3*x1 + w4*x2 + b2)
    o=sigmoid(w5*h1+w6*h2+b3)

    print(round(o))
```
