```python
import numpy as np

def max_min_composition(R, S):
    # R is m x n matrix, S is n x p matrix
    # Result T is m x p matrix
    m = R.shape[0]
    p = S.shape[1]
    n = R.shape[1]
    
    # Initialize result matrix
    T = np.zeros((m, p))
    
    for i in range(m):
        for j in range(p):
            # For each element (i, j) in T, 
            # find the max-min over intermediate y (n)
            min_values = []
            for k in range(n):
                min_values.append(min(R[i, k], S[k, j]))
            T[i, j] = max(min_values)
            
    return T

# Example Input Matrices (Fuzzy Relations)
# R is 2x2, S is 2x3
R = np.array([[0.7, 0.4],
              [0.9, 0.2]])
S = np.array([[0.6, 0.8, 0.2],
              [0.5, 0.3, 0.9]])

# Calculate Composition
result = max_min_composition(R, S)
print("Result of Max-Min Composition:\n", result)
```
